```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
celltype = 'spermatocyte'
network = 'thr'
layer = 'HS'
threshold = 0.5
threshold_str = str(threshold).replace('.', 'p')
```

```python
rPosTxtFile = 'results/forceatlas2/{celltype:s}/net-{celltype:s}-{network:s}-{threshold:s}-{layer:s}-forceatlas2.txt'.format(celltype=celltype, network=network, threshold=threshold_str, layer=layer)
```

```python
df = pd.read_csv(rPosTxtFile, sep='\t')
df['color'] = 'blue'
df.loc[df['id']=='ENSG00000186272','color'] = 'red'
```

```python
fig, ax = plt.subplots(figsize=(12,12))
df.plot(x='x', y='y', c='color', kind='scatter', grid=True, ax=ax)
ax.plot([xl,xl,xh,xh,xl], [yl,yh,yh,yl,yl], color='green')
```
